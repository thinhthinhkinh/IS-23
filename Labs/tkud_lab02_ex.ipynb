## NUMPY

Bạn hãy phát sinh ngẫu nhiên mảng ndarray có tên `arr` gồm 100 phần tử số nguyên có giá trị từ 0..1000. Thực hiện các thao tác sau:

### 1. Hiển thị các phần tử thỏa yêu cầu dưới đây

- 10 phần tử đầu tiên
- 5 phần tử cuối cùng
- Các phần tử thỏa mãn điều kiện chia hết cho 5 và lớn hơn 100

```python
import numpy as np

a = np.random.randint(0, 1000,size=100)
print(a[:10])
print(a[-5:])
```

### 2. Cho biết mảng trên có bao nhiêu phần tử có giá trị nhỏ hơn 100. Hiển thị các phần tử đó

```python
import numpy as np

a = np.random.randint(0, 1000,size=100)
print(a[a<100])
```

### 3. Thực hiện thống kê:

- Tính tổng mảng
- Tìm giá trị trung bình
- Tìm giá trị trung vị
- Tìm Q1, Q2, Q3
- Tìm phân vị 20 (percentile 20)

```python
import numpy as np

a = np.random.randint(0,1000, size=1000)

tong=np.sum(a)
print(tong)

trungbinh=np.mean(a)
print(trungbinh)

trungvi=np.median(a)
print(trungvi)

q1=np.percentile(a,25)
print(q1)

q2=np.percentile(a,50)
print(q2)

q3=np.percentile(a,75)
print(q3)
print(f"Q1: {q1}, Q2: {q2}, Q3: {q3}")

phanvi=np.percentile(a,20)
print(f"Phan vi: {phanvi}")
```

### 4. Thực hiện việc chọn ngẫu nhiên (không thay thế) 10 phần tử từ mảng trên 5 lần. Hiển thị kết quả lựa chọn mỗi lần

```python

import numpy as np

a = np.random.randint(0,1000, size=1000)

for i in range(5):
    print(np.random.choice(a,10))
```

### 5. Tìm phần tử xuất hiện nhiều nhất trong mảng

```python
import numpy as np

a = np.random.randint(0,1000, size=1000)

print(np.bincount(a).argmax())
```

### Câu 4: Định dạng đường thẳng

Phát sinh x là các điểm dữ liệu từ $[0, \pi]$. Bạn hãy vẽ đồ thị $y = sin(x)$ và $y = cos(x)$ như dưới đây.
Gợi ý: sử dụng hằng số pi và hàm sau:
- `np.pi` : số $\pi$
- `np.linspace()`
- `np.sin()`
- `np.cos()`

```python
# This cell contains output. Don't remove!
```

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(0, 2 * np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize=(10, 5))
plt.plot(x, y_sin, label='sin(x)', color='blue')
plt.plot(x, y_cos, label='cos(x)', color='orange', linestyle='--')

plt.title('Đồ thị hàm sin(x) và cos(x)')


plt.grid(True)
plt.legend()

plt.show()
```

### Câu 8: Vẽ histogram

Sử dụng hàm `np.random.normal(size=10000)` để phát sinh 10000 số ngẫu nghiên thuộc phân phối chuẩn. Vẽ histogram các số ngẫu nhiên trên.

```python
# This cell contains output. Don't remove!
```

```python
import numpy as np
import matplotlib.pyplot as plt


data = np.random.normal(size=10000)

plt.figure(figsize=(10, 6))
plt.hist(data, bins=50)


plt.title('Histogram', fontsize=14)
plt.xlabel('Value', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

plt.xlim(-4, 4)

plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

### Câu 9: Vẽ đồ thị hộp

Sử dụng hàm `np.random.normal(size=10000)` để phát sinh 10000 số ngẫu nghiên thuộc phân phối chuẩn. Vẽ boxplot các số ngẫu nhiên trên.

```python
# This cell contains output. Don't remove!
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(size=10000)

plt.figure(figsize=(8, 6))
bp = plt.boxplot(data, patch_artist=True, vert=True)

plt.setp(bp['boxes'], facecolor='white', edgecolor='black')
plt.setp(bp['medians'], color='orange')
plt.setp(bp['whiskers'], color='black')
plt.setp(bp['caps'], color='black')
plt.setp(bp['fliers'], marker='o', markerfacecolor='black', markersize=3)

plt.title('Histogram', fontsize=14)
plt.xlabel('Value', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

plt.ylim(-4, 4)


plt.xticks([1], ['1'])

plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Câu 10: Kết hợp trên cùng một đồ thị

Sử dụng hàm `np.random.normal(size=10000)` để phát sinh 10000 số ngẫu nghiên thuộc phân phối chuẩn. Vẽ hình thể hiện phân phối như sau:

Hướng dẫn: thêm tham số `vert=False` trong hàm vẽ `boxplot()`

```python
# This cell contains output. Don't remove!
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(size=10000)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8), sharex=True, gridspec_kw={'height_ratios': [3, 1]})

ax1.hist(data, bins=50, density=True, alpha=0.7, color='purple')
ax1.set_ylabel('Mật độ')
ax1.set_title('Đồ thị hàm sin(x) và cos(x)\nHistogram')

bp = ax2.boxplot(data, vert=False, patch_artist=True)
ax2.set_ylabel('Boxplot')
ax2.set_xlabel('Giá trị')

plt.setp(bp['boxes'], facecolor='white', edgecolor='black')
plt.setp(bp['medians'], color='orange')
plt.setp(bp['whiskers'], color='black')
plt.setp(bp['caps'], color='black')
plt.setp(bp['fliers'], marker='o', markerfacecolor='black', markersize=3)


plt.xlim(-4, 4)


plt.tight_layout()
plt.show()
```
