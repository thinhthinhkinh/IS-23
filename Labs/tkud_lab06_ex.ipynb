### Bài 1: Tạo quần thể mô phỏng

Bạn hãy tạo một mẫu gồm 10000 phần tử mô phỏng chiều cao của nam thanh niên có giá trị từ 120cm - 200cm. Bạn hãy lưu kết quả vào biến **POP**.

Tính:
a. Chiều cao trung bình (kỳ vọng) của quần thể và độ lệch chuẩn về chiều cao của quần thể.

b. Tính tỷ lệ người cao trong quần thể, biết rằng thanh niên có chiều cao từ 180 trở lên được gọi là cao.

c. Vẽ histogram về chiều cao của quần thể. Theo bạn quần thể có phân phối chuẩn hay không?

d. Thử vẽ histogram và đồ thị hàm mật độ của phân phối chuẩn sử dụng tham số loc và scale bằng với kỳ vọng và độ lệch chuẩn của quần thể.

```python
import numpy as np
import scipy.stats as stats

np.random.seed(123)
POP = np.random.uniform(120, 200, 10000)
mean_height = np.mean(POP)  # Chiều cao trung bình
sd_height = np.std(POP)      # Độ lệch chuẩn

print(f"Chiều cao trung bình: {mean_height:.2f} cm")
print(f"Độ lệch chuẩn: {sd_height:.2f} cm")
tall_ratio = np.sum(POP >= 180) / len(POP)  # Tỷ lệ người cao (chiều cao >= 180cm)
print(f"Tỷ lệ người cao: {tall_ratio:.2%}")
import matplotlib.pyplot as plt

plt.hist(POP, bins=30, color='lightblue', edgecolor='black')
plt.title('Histogram of Heights')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.show()

stats.probplot(POP, dist="norm", plot=plt)
plt.title('QQ Plot')
plt.show()
# Vẽ histogram
plt.hist(POP, bins=30, density=True, color='lightblue', edgecolor='black', alpha=0.6)
plt.title('Histogram with Normal Density')
plt.xlabel('Height (cm)')
plt.ylabel('Density')

# Thêm hàm mật độ chuẩn
x = np.linspace(min(POP), max(POP), 100)
y = stats.norm.pdf(x, mean_height, sd_height)
plt.plot(x, y, color='red', linewidth=2)
plt.show()

```

### Bài 4: Khảo sát phân phối của trung bình mẫu

a. Thực hiện bài 2 100 lần, mỗi lần bạn tính được trung bình mẫu. Vẽ đồ thị histogram cho 100 trung bình mẫu bạn tính được.

b. Thực hiện bài 3 100 lần, mỗi lần bạn tính được trung bình mẫu. Vẽ đồ thị histogram cho 100 trung bình mẫu bạn tính được.

c. Vẽ histogram của hai câu a, b trong cùng một hình để so sánh. Theo bạn hình dạng của histogram thay đổi nói lên điều gì?

Gợi ý: tạo một danh sách rỗng để chứa các trung bình mẫu tính được, sử dụng hàm append() để thêm giá trị trung bình vào danh sách sau mỗi lần tính.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(123)  # Đặt seed để có thể tái tạo kết quả
sample_size_1 = 20
num_trials = 100
means_20 = []

for _ in range(num_trials):
    sample = np.random.uniform(120, 200, sample_size_1)
    means_20.append(np.mean(sample))

# Vẽ histogram cho 100 trung bình mẫu
plt.hist(means_20, bins=10, color='lightblue', edgecolor='black')
plt.title('Histogram of Sample Means (n=20)')
plt.xlabel('Mean Height (cm)')
plt.ylabel('Frequency')
plt.show()
sample_size_2 = 100
means_100 = []

for _ in range(num_trials):
    sample = np.random.uniform(120, 200, sample_size_2)
    means_100.append(np.mean(sample))

# Vẽ histogram cho 100 trung bình mẫu
plt.hist(means_100, bins=10, color='lightgreen', edgecolor='black')
plt.title('Histogram of Sample Means (n=100)')
plt.xlabel('Mean Height (cm)')
plt.ylabel('Frequency')
plt.show()
plt.hist(means_20, bins=10, color='lightblue', edgecolor='black', alpha=0.5, label='n=20')
plt.hist(means_100, bins=10, color='lightgreen', edgecolor='black', alpha=0.5, label='n=100')
plt.title('Comparison of Sample Means Histograms')
plt.xlabel('Mean Height (cm)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

### Bài 5: Khảo sát tỷ lệ mẫu

### Bài 6: Khảo sát phương sai mẫu

Bạn hãy mô phỏng 1000 lần lấy mẫu có kích cỡ 100. Vẽ histogram đại lượng phương sai của các mẫu (Phương sai bằng bình phương độ lệch chuẩn)

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(123)  # Đặt seed để có thể tái tạo kết quả
sample_size = 100
num_trials = 1000
variances = []

for _ in range(num_trials):
    sample = np.random.uniform(120, 200, sample_size)
    variance = np.var(sample, ddof=1)  # Tính phương sai mẫu
    variances.append(variance)

# Vẽ histogram phương sai mẫu
plt.hist(variances, bins=10, color='lightblue', edgecolor='black')
plt.title('Histogram of Sample Variances (n=100)')
plt.xlabel('Variance')
plt.ylabel('Frequency')
plt.show()
```

### Bài 7:  Bài Tập Ước Lượng

Khảo sát cân nặng (kg) của gà khi xuất chuồng, người ta cân một số con và có kết quả như sau:

> 2.1, 1.8, 2.0, 2.3, 1.7, 1.5, 2.0, 2.2, 1.8

Giả sử cân nặng của gà là biến ngẫu nhiên có phân phối chuẩn. Hãy lập khoảng tin cậy cho cân nặng trung bình của  khi xuất chuồng với độ tin cậy là 95%. Trong 2 trường hợp:

a, Biết $\sigma = 0.3$

b, Không biết $\sigma$


```python
import numpy as np
import scipy.stats as stats

# Dữ liệu cân nặng của gà
weights = np.array([2.1, 1.8, 2.0, 2.3, 1.7, 1.5, 2.0, 2.2, 1.8])

# Kích thước mẫu
n = len(weights)

# Tính toán trung bình mẫu
mean_weight = np.mean(weights)

# a. Biết σ = 0.3
sigma = 0.3
z_score = stats.norm.ppf(0.975)  # Z cho 95% tin cậy

margin_of_error_known_sigma = z_score * (sigma / np.sqrt(n))
confidence_interval_known_sigma = (mean_weight - margin_of_error_known_sigma, mean_weight + margin_of_error_known_sigma)

print("Khoảng tin cậy cho trung bình (biết σ):", confidence_interval_known_sigma)

# b. Không biết σ
s = np.std(weights, ddof=1)  # Độ lệch chuẩn mẫu
t_score = stats.t.ppf(0.975, df=n-1)  # t cho 95% tin cậy

margin_of_error_unknown_sigma = t_score * (s / np.sqrt(n))
confidence_interval_unknown_sigma = (mean_weight - margin_of_error_unknown_sigma, mean_weight + margin_of_error_unknown_sigma)

print("Khoảng tin cậy cho trung bình (không biết σ):", confidence_interval_unknown_sigma)
```

### Bài 8

Bộ dữ liệu `SAT-lard.csv` chứa điểm thi SAT của 1000 học sinh.

b. Giả sử bộ dữ liệu là kết quả của quá trình lấy mẫu ngẫu nhiên từ các học sinh. Bạn hãy xây dựng khoảng tin cậy cho trung bình điểm thi của tất cả các sinh viên với độ tin cậy (`Đáp án: 105.2 ± 1.86`)

d. Giả sử bộ dữ liệu là toàn bộ kết quả điểm thi SAT của học sinh một trường phố thông. Sử dụng mẫu là điểm thi của 36 học sinh đầu tiên trong bộ dữ liệu, bạn hãy xây dựng khoảng tin cậy **99%** cho điểm SAT trung bình của trường phổ thông đấy. (`Đáp án: (1428.22, 1602.89)`)

e. Giả sử bộ dữ liệu là toàn bộ kết quả điểm thi SAT của học sinh một trường phố thông. Sử dụng mẫu là điểm thi của 36 học sinh đầu tiên trong bộ dữ liệu, bạn hãy xây dựng khoảng tin cậy **95%** cho điểm SAT trung bình của trường phổ thông đấy. Cho biết khoảng ước lượng bạn xây dựng có đúng không, dựa trên giá trị đã tính ở câu trên.

```python
import pandas as pd
import numpy as np
import scipy.stats as stats

# Đọc dữ liệu từ file CSV
path = "/content/drive/MyDrive/data/SAT-lard.csv"
data = pd.read_csv(path)

# In ra danh sách các cột để kiểm tra tên cột
# print(data.columns) # Kết quả in ra cho thấy tên cột là 'SAT Score' và 'College GPA'
# Do đó, cần thay 'score' thành 'SAT Score' trong code

# Tính toán các thông số cần thiết
mean_score = data['SAT Score'].mean()  # Thay 'score' bằng 'SAT Score'
std_dev = data['SAT Score'].std(ddof=1)
n = len(data)

# Khoảng tin cậy 99%
confidence_level = 0.99
z_score = stats.norm.ppf((1 + confidence_level) / 2)

margin_of_error = z_score * (std_dev / np.sqrt(n))
confidence_interval_99 = (mean_score - margin_of_error, mean_score + margin_of_error)

print("Khoảng tin cậy 99%:", confidence_interval_99)
# Khoảng tin cậy 90%
confidence_level = 0.90
z_score = stats.norm.ppf((1 + confidence_level) / 2)

margin_of_error_90 = z_score * (std_dev / np.sqrt(n))
confidence_interval_90 = (mean_score - margin_of_error_90, mean_score + margin_of_error_90)

print("Khoảng tin cậy 90%:", confidence_interval_90)
# Tính điểm SAT trung bình
overall_mean = data['SAT Score'].mean() # Thay 'score' bằng 'SAT Score'
print("Điểm SAT trung bình của trường phổ thông:", overall_mean)
# Sử dụng mẫu 36 học sinh đầu tiên
sample = data['SAT Score'].head(36) # Thay 'score' bằng 'SAT Score'
sample_mean = sample.mean()
sample_std_dev = sample.std(ddof=1)
sample_n = len(sample)

# Khoảng tin cậy 99% cho mẫu
confidence_level = 0.99
t_score = stats.t.ppf((1 + confidence_level) / 2, df=sample_n - 1)

margin_of_error_sample = t_score * (sample_std_dev / np.sqrt(sample_n))
confidence_interval_sample_99 = (sample_mean - margin_of_error_sample, sample_mean + margin_of_error_sample)

print("Khoảng tin cậy 99% cho mẫu 36 học sinh đầu tiên:", confidence_interval_sample_99)
# Khoảng tin cậy 95% cho mẫu
confidence_level = 0.95
t_score_95 = stats.t.ppf((1 + confidence_level) / 2, df=sample_n - 1)

margin_of_error_sample_95 = t_score_95 * (sample_std_dev / np.sqrt(sample_n))
confidence_interval_sample_95 = (sample_mean - margin_of_error_sample_95, sample_mean + margin_of_error_sample_95)

print("Khoảng tin cậy 95% cho mẫu 36 học sinh đầu tiên:", confidence_interval_sample_95)
```
