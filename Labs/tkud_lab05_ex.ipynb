**Bài 01:** Một bài thi trắc nghiệm gồm 10 câu hỏi, mỗi câu có 4 phương án trả lời trong đó chỉ có một phương án đúng. Bạn hãy:

a. Một sinh viên không học bài làm bài bằng cách chọn ngẫu nhiên một phương án cho mỗi câu hỏi. Bạn hãy dùng hàm random để in ra số câu đúng của bạn sinh viên trong 10 lần kiểm tra.

b. Giả sử mỗi câu đúng được 4 điểm, mỗi câu sai bị trừ 2 điểm. Tính xác suất để sinh viên này được 4 điểm

c. Gọi X là số câu trả lời đúng. Tính E(X) và Var(X)

```python
import random

# Phần a: Mô phỏng 10 bài kiểm tra và in ra số câu đúng của mỗi bài kiểm tra
def mo_phong_bai_kiem_tra(so_cau_hoi=10):
  """Mô phỏng một bài kiểm tra trắc nghiệm và trả về số câu trả lời đúng."""
  cau_tra_loi_dung = 0
  for _ in range(so_cau_hoi):
    # Giả sử 1 đại diện cho câu trả lời đúng và 2, 3, 4 là sai
    if random.choice([1, 2, 3, 4]) == 1:
      cau_tra_loi_dung += 1
  return cau_tra_loi_dung

print("Số câu trả lời đúng trong 10 bài kiểm tra:")
for _ in range(10):
  print(mo_phong_bai_kiem_tra())


# Phần b: Tính xác suất để đạt được điểm số 4
def tinh_xac_suat_diem_4(so_cau_hoi=10):
  """Tính xác suất đạt được điểm số 4 trong bài kiểm tra."""
  tong_so_ket_qua = 4**so_cau_hoi  # Tổng số kết quả có thể có
  ket_qua_thuan_loi = 0  # Khởi tạo số lượng kết quả dẫn đến điểm số 4

  # Lặp qua tất cả các kết quả có thể có
  for i in range(tong_so_ket_qua):
    cau_tra_loi_dung = 0
    temp = i
    for _ in range(so_cau_hoi):
      if temp % 4 == 0:  # Giả sử 0 đại diện cho câu trả lời đúng cho phép tính này
        cau_tra_loi_dung += 1
      temp //= 4

    diem_so = cau_tra_loi_dung * 4 - (so_cau_hoi - cau_tra_loi_dung) * 2
    if diem_so == 4:
      ket_qua_thuan_loi += 1

  return ket_qua_thuan_loi / tong_so_ket_qua

xac_suat = tinh_xac_suat_diem_4()
print(f"\nXác suất đạt được điểm số 4: {xac_suat}")
```

**Bài 02:** Một trung tâm bưu điện nhận được 3 cuộc gọi điện thoại mỗi phút. Tính xác suất để trung tâm nhận được 1 cuộc gọi, 2 cuộc gọi, 3 cuộc gọi trong vòng một phút biết số cuộc gọi trong một phút có phân phối Poisson.

```python
import math

def xac_suat_poisson(k, lamda):
  """Tính toán xác suất của k sự kiện xảy ra trong một khoảng thời gian nhất định
  với tốc độ trung bình là lamda."""
  return (lamda**k * math.exp(-lamda)) / math.factorial(k)

# Tốc độ trung bình của cuộc gọi (lamda) là 3 cuộc gọi mỗi phút
lamda = 3

# Tính toán xác suất cho 1, 2 và 3 cuộc gọi
xac_suat_1_cuoc_goi = xac_suat_poisson(1, lamda)
xac_suat_2_cuoc_goi = xac_suat_poisson(2, lamda)
xac_suat_3_cuoc_goi = xac_suat_poisson(3, lamda)

# In kết quả
print(f"Xác suất nhận được 1 cuộc gọi: {xac_suat_1_cuoc_goi}")
print(f"Xác suất nhận được 2 cuộc gọi: {xac_suat_2_cuoc_goi}")
print(f"Xác suất nhận được 3 cuộc gọi: {xac_suat_3_cuoc_goi}")
```

**Bài 03:** Trọng lượng (đơn vị gam) của một loại trái cây có phân phối chuẩn với µ = 500 (gam) gam và $σ^2 = 16 (gam^2)$. Trái cây thu hoạch được phân loại theo trọng lượng như sau:

Loại 1: trên 505 gam

Loại 2: từ 495 – 505 gam

Loại 3: dười 495 gam

a. Hãy tính tỷ lệ của mỗi loại.

d. Từ hai cách lấy mẫu ở câu b, c bạn có nhận xét gì không?

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Phần a: Tính tỷ lệ của mỗi loại trái cây
mu = 500
sigma = np.sqrt(16)  # Độ lệch chuẩn

# Tính toán xác suất bằng cách sử dụng hàm phân phối tích lũy (cdf)
ty_le_loai_1 = 1 - stats.norm.cdf(505, loc=mu, scale=sigma)
ty_le_loai_2 = stats.norm.cdf(505, loc=mu, scale=sigma) - stats.norm.cdf(495, loc=mu, scale=sigma)
ty_le_loai_3 = stats.norm.cdf(495, loc=mu, scale=sigma)

print(f"Tỷ lệ loại 1: {ty_le_loai_1}")
print(f"Tỷ lệ loại 2: {ty_le_loai_2}")
print(f"Tỷ lệ loại 3: {ty_le_loai_3}")

# Phần b: Mô phỏng lấy mẫu 20 trái cây, vẽ histogram và tính toán thống kê
kich_thuoc_mau = 20
trong_luong_trai_cay = np.random.normal(mu, sigma, kich_thuoc_mau)

# Vẽ histogram
plt.hist(trong_luong_trai_cay, bins=10, edgecolor='black')
plt.xlabel("Trọng lượng (gam)")
plt.ylabel("Tần số")
plt.title("Histogram Trọng lượng Trái cây (Mẫu 20)")
plt.show() #Remember to execute this code yourself to see the output

# Tính toán giá trị trung bình và độ lệch chuẩn của mẫu
gia_tri_trung_binh_mau_20 = np.mean(trong_luong_trai_cay)
do_lech_chuan_mau_20 = np.std(trong_luong_trai_cay)

print(f"Giá trị trung bình (mẫu 20): {gia_tri_trung_binh_mau_20}")
print(f"Độ lệch chuẩn (mẫu 20): {do_lech_chuan_mau_20}")

# Phần c: Mô phỏng lấy mẫu 100 trái cây, vẽ histogram và tính toán thống kê
kich_thuoc_mau = 100
trong_luong_trai_cay = np.random.normal(mu, sigma, kich_thuoc_mau)

# Vẽ histogram
plt.hist(trong_luong_trai_cay, bins=10, edgecolor='black')
plt.xlabel("Trọng lượng (gam)")
plt.ylabel("Tần số")
plt.title("Histogram Trọng lượng Trái cây (Mẫu 100)")
plt.show() #Remember to execute this code yourself to see the output

# Tính toán giá trị trung bình và độ lệch chuẩn của mẫu
gia_tri_trung_binh_mau_100 = np.mean(trong_luong_trai_cay)
do_lech_chuan_mau_100 = np.std(trong_luong_trai_cay)

print(f"Giá trị trung bình (mẫu 100): {gia_tri_trung_binh_mau_100}")
print(f"Độ lệch chuẩn (mẫu 100): {do_lech_chuan_mau_100}")

# Phần d: Nhận xét
print("\nNhận xét:")
print("- Khi kích thước mẫu tăng từ 20 lên 100, histogram có xu hướng giống với phân phối chuẩn hơn.")
print("- Giá trị trung bình và độ lệch chuẩn của mẫu có xu hướng tiệm cận với giá trị trung bình và độ lệch chuẩn của quần thể khi kích thước mẫu tăng.")
```

**Bài 04:** Mô phỏng tung một con xúc sắc cân đối đồng chất 5000 lần. Dựa vào giá trị mô phỏng, bạn hãy tìm các giá trị xác suất dưới đây:

a. Xác suất để  số chấm xuất hiện là 4

b. Xác suất để số chấm xuất hiện lớn hơn hoặc bằng 4

c. Giả sử biết số chấm xuất hiện lớn hơn hoặc bằng 4. Hãy tìm xác suất để mặt 6 chấm xuất hiện.

```python
import random

num_trials = 5000
dice_rolls = [random.randint(1, 6) for _ in range(num_trials)]

# a. Xác suất để số chấm xuất hiện là 4
count_4 = dice_rolls.count(4)
prob_4 = count_4 / num_trials
print(f"a. Xác suất xuất hiện mặt 4: {prob_4}")

# b. Xác suất để số chấm xuất hiện lớn hơn hoặc bằng 4
count_greater_equal_4 = sum(1 for roll in dice_rolls if roll >= 4)
prob_greater_equal_4 = count_greater_equal_4 / num_trials
print(f"b. Xác suất xuất hiện mặt lớn hơn hoặc bằng 4: {prob_greater_equal_4}")

# c. Xác suất xuất hiện mặt 6, biết số chấm lớn hơn hoặc bằng 4
count_6_given_greater_equal_4 = sum(1 for roll in dice_rolls if roll == 6 and roll >= 4)
prob_6_given_greater_equal_4 = count_6_given_greater_equal_4 / count_greater_equal_4
print(f"c. Xác suất xuất hiện mặt 6, biết số chấm lớn hơn hoặc bằng 4: {prob_6_given_greater_equal_4}")
```

**Bài 05:** Tạo 5000 số ngẫu nhiên có phân phối nhị thức với n=50, p=0.7 Bạn Hãy

a. Tìm các giá trị thống kê: min, max, Q1, Q2, Q3

b. Vẽ đồ thị boxplot

c. Vẽ các đồ thị: tần số, tần suất

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate 5000 random numbers from binomial distribution
n = 50
p = 0.7
random_numbers = np.random.binomial(n, p, size=5000)

# a. Calculate statistical values
min_val = np.min(random_numbers)
max_val = np.max(random_numbers)
q1 = np.quantile(random_numbers, 0.25)
q2 = np.quantile(random_numbers, 0.50)  # Median
q3 = np.quantile(random_numbers, 0.75)

print(f"Min: {min_val}")
print(f"Max: {max_val}")
print(f"Q1: {q1}")
print(f"Q2 (Median): {q2}")
print(f"Q3: {q3}")

# b. Create a boxplot
plt.figure()
sns.boxplot(x=random_numbers)
plt.title("Boxplot của Số Ngẫu Nhiên")
plt.xlabel("Giá Trị")
plt.show() #Remember to execute this code yourself to see the output


# c. Create frequency and relative frequency histograms
plt.figure()
plt.hist(random_numbers, bins=20, edgecolor='black')
plt.title("Biểu đồ Tần số")
plt.xlabel("Giá Trị")
plt.ylabel("Tần số")
plt.show() #Remember to execute this code yourself to see the output


plt.figure()
plt.hist(random_numbers, bins=20, edgecolor='black', density=True)  # density=True for relative frequency
plt.title("Biểu đồ Tần suất")
plt.xlabel("Giá Trị")
plt.ylabel("Tần suất")
plt.show() #Remember to execute this code yourself to see the output
```

$P(X \leq 5), P(X \le 5), P(X \ge 4), P(X = 5) $

```python
from scipy.stats import binom, poisson

# a. Ý nghĩa của binom.pmf(4, 10, 0.6)
print("binom.pmf(4, 10, 0.6): tính xác suất để biến ngẫu nhiên nhị thức X ~ B(10, 0.6) nhận giá trị bằng 4.")
print("Nói cách khác, nó tính P(X = 4) với X ~ B(10, 0.6).")

# b. Tính toán xác suất cho phân phối nhị thức X ~ B(10, 0.6)
print("\nb. Xác suất cho phân phối nhị thức X ~ B(10, 0.6):")
print(f"P(X ≤ 5) = {binom.cdf(5, 10, 0.6)}")  # Cumulative distribution function
print(f"P(X ≤ 5) = {binom.cdf(5, 10, 0.6)}")  # Same as above
print(f"P(X ≥ 4) = {1 - binom.cdf(3, 10, 0.6)}")  # 1 - P(X ≤ 3)
print(f"P(X = 5) = {binom.pmf(5, 10, 0.6)}")  # Probability mass function


# c. Ý nghĩa của poisson.pmf(4, 3)
print("\nc. poisson.pmf(4, 3): tính xác suất để biến ngẫu nhiên Poisson X ~ P(3) nhận giá trị bằng 4.")
print("Nói cách khác, nó tính P(X = 4) với X ~ P(3).")

# d. Tính toán xác suất cho phân phối Poisson X ~ P(3)
print("\nd. Xác suất cho phân phối Poisson X ~ P(3):")
print(f"P(X ≤ 5) = {poisson.cdf(5, 3)}")
print(f"P(X ≤ 5) = {poisson.cdf(5, 3)}")  # Same as above
print(f"P(X ≥ 4) = {1 - poisson.cdf(3, 3)}")  # 1 - P(X ≤ 3)
print(f"P(X = 5) = {poisson.pmf(5, 3)}")
```

Một công ty sản xuất làm ra các chi tiết máy với độ dài các chi tiết có phân phối chuẩn với kỳ vọng là 3000 mm và độ lệch chuẩn là 3 mm. Yêu cầu cho các chi tiết máy này là phải có chiều dài từ 2993 mm đến 3007 mm. Tính tỷ lệ sản phẩm lỗi của công ty

**Bài 10**

Viết một chương trình để mô phỏng việc tung đồng xu và đánh giá phương sai của giá trị mô phỏng của biến ngẫu nhiên X (tung được mặt `Head`) đối với các số lần tung khác nhau. Giả sử  đồng xu phải cân bằng.

Đối với mỗi N từ 10; 40; 90; 160; 250; 490; 640; 810; 1000, hãy ước tính giá trị của X bằng cách mô phỏng số lần tung đó.

Bạn nên chạy mỗi mô phỏng 100 lần và sử dụng tập hợp các ước tính để đánh giá phương sai của ước tính x của bạn. Vẽ biểu đồ phương sai này so với N, bạn có nhận xét gì không?

```python
import numpy as np
import matplotlib.pyplot as plt

def mo_phong_tung_dong_xu(N, so_lan_mo_phong=100):
    """Mô phỏng việc tung đồng xu và ước tính phương sai của tỷ lệ xuất hiện mặt ngửa."""
    ty_le_mat_ngua = []
    for _ in range(so_lan_mo_phong):
        # Mô phỏng N lần tung đồng xu (0 cho mặt sấp, 1 cho mặt ngửa)
        ket_qua_tung = np.random.randint(0, 2, N)
        # Tính toán tỷ lệ mặt ngửa
        ty_le = np.sum(ket_qua_tung) / N
        ty_le_mat_ngua.append(ty_le)

    # Tính toán phương sai của tỷ lệ mặt ngửa
    phuong_sai = np.var(ty_le_mat_ngua)
    return phuong_sai

# Danh sách các giá trị N
cac_gia_tri_N = [10, 40, 90, 160, 250, 490, 640, 810, 1000]

# Tính toán phương sai cho mỗi N, chạy 100 lần mô phỏng cho mỗi N
phuong_sai = [mo_phong_tung_dong_xu(N, so_lan_mo_phong=100) for N in cac_gia_tri_N]

# Vẽ biểu đồ kết quả
plt.plot(cac_gia_tri_N, phuong_sai, 'o-')
plt.xlabel("Số lần tung (N)")
plt.ylabel("Phương sai")
plt.title("Phương sai của Tỷ lệ Mặt Ngửa theo Số lần Tung (100 lần mô phỏng)")
plt.show()

print("Nhận xét:")
print("Khi số lần tung (N) tăng lên, phương sai của tỷ lệ mặt ngửa giảm xuống.")
print("Điều này phù hợp với quy luật số lớn, cho biết khi kích thước mẫu tăng, giá trị trung bình mẫu sẽ tiệm cận với giá trị kỳ vọng (0.5 cho đồng xu cân bằng).")
print("Phương sai giảm phản ánh sự hội tụ này và cho thấy ước tính tỷ lệ mặt ngửa trở nên chính xác hơn khi số lần tung tăng lên.")
```
